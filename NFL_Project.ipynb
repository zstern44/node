```
chosen_team = input("""With which team are you concerned?
Enter one of the following abbreviations exactly as written below:
ARI, ATL, BAL, BUF, CAR, CHI, CIN, CLE,
DAL, DEN, DET, GB, HOU, IND, JAX, KC,
LAC, LAR, MIA, MIN, NE, NO, NYG, NYJ,
OAK, PHI, PIT, SEA, SF, TB, TEN, WAS

Your chosen team is: """)


intercept_coefficient = -0.31161
DEF_TAKEAWAYS_GAME_coefficient = 1.903169
OFF_GIVEAWAYS_GAME_coefficient = -3.16506
DEF_SACKS_GAME_coefficient = 0.607806
OPP_YARDS_RUSH_coefficient = -0.73712
OPP_YARDS_PASSATT_coefficient = -1.21027
PASS_ATTEMPTS_GAME_coefficient = 0.136455
RUSH_ATTEMPTS_GAME_coefficient = 0.204057
YARDS_PER_PASS_ATT_coefficient = 1.610765

def stats_chosen_team(chosen_team):
	matching_rows = nfl_df[(nfl_df["YEAR"] == 2019) & (nfl_df["Team"] == chosen_team)]

	print(f"Here are {chosen_team}'s 2019 stats: ")
	print("Defense Takeaways Per Game: ", matching_rows["DEF_TAKEAWAYS_GAME"].values)
	print("Offensive Giveaways Per Game: ", matching_rows["OFF_GIVEAWAYS_GAME"].values)
	print("Defensive Sacks Per Game: ", matching_rows["DEF_SACKS_GAME"].values)
	print("Opponents' Yards Per Rush: ", matching_rows["OPP_YARDS_RUSH"].values)
	print("Opponent's Yards per Pass Attempts: ", matching_rows["OPP_YARDS_PASSATT"].values)
	print("Pass Attempts Per Game: ", matching_rows["PASS_ATTEMPTS_GAME"].values)
	print("Rush Attempts Per Game: ", matching_rows["RUSH_ATTEMPTS_GAME"].values)
	print("Yards Per Pass Attempt: ", matching_rows["YARDS_PER_PASS_ATT"].values)


stats_chosen_team(chosen_team)

print("""
""")

def average_stats_2013_2019(nfl_df):
	average_DEF_TAKEAWAYS_GAME = sum(nfl_df["DEF_TAKEAWAYS_GAME"]) / 224
	average_OFF_GIVEAWAYS_GAME = sum(nfl_df["OFF_GIVEAWAYS_GAME"]) / 224
	average_DEF_SACKS_GAME = sum(nfl_df["DEF_SACKS_GAME"]) / 224
	average_OPP_YARDS_RUSH = sum(nfl_df["OPP_YARDS_RUSH"]) / 224
	average_OPP_YARDS_PASSATT = sum(nfl_df["OPP_YARDS_PASSATT"]) / 224
	average_PASS_ATTEMPTS_GAME = sum(nfl_df["PASS_ATTEMPTS_GAME"]) / 224
	average_RUSH_ATTEMPTS_GAME = sum(nfl_df["RUSH_ATTEMPTS_GAME"]) / 224
	average_YARDS_PER_PASS_ATT = sum(nfl_df["YARDS_PER_PASS_ATT"]) / 224

	print("The Following are the league averages for each stat between 2013 and 2019: ")
	print("League Average Defensive Takeaways Per Game: ", average_DEF_TAKEAWAYS_GAME)
	print("League Average Offensive Giveaways Per Game: ", average_OFF_GIVEAWAYS_GAME)
	print("League Average Defensive Sacks Per Game: ", average_DEF_SACKS_GAME)
	print("League Average Opponents' Yards Per Rush: ", average_OPP_YARDS_RUSH)
	print("League Average Opponents' Yards Per Pass Attempts: ", average_OPP_YARDS_PASSATT)
	print("League Average Pass Attempts Per Game: ", average_PASS_ATTEMPTS_GAME)
	print("League Average Rush Attempts Per Game: ", average_RUSH_ATTEMPTS_GAME)
	print("League Average Yards Per Pass Attempts: ", average_YARDS_PER_PASS_ATT)

average_stats_2013_2019(nfl_df)

print("""
""")

def projected_wins_2020(chosen_team):
	matching_rows = nfl_df[(nfl_df["YEAR"] == 2019) & (nfl_df["Team"] == chosen_team)]
	
	DEF_TAKEAWAYS_GAME = matching_rows["DEF_TAKEAWAYS_GAME"]
	OFF_GIVEAWAYS_GAME = matching_rows["OFF_GIVEAWAYS_GAME"]
	DEF_SACKS_GAME = matching_rows["DEF_SACKS_GAME"]
	OPP_YARDS_RUSH = matching_rows["OPP_YARDS_RUSH"]
	OPP_YARDS_PASSATT = matching_rows["OPP_YARDS_PASSATT"]
	PASS_ATTEMPTS_GAME = matching_rows["PASS_ATTEMPTS_GAME"]
	RUSH_ATTEMPTS_GAME = matching_rows["RUSH_ATTEMPTS_GAME"]
	YARDS_PER_PASS_ATT = matching_rows["YARDS_PER_PASS_ATT"]

	projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	predicted_2020_wins_for_team = projected_wins_equation.values

	return f"If {chosen_team} keeps the same stats for 2020, it should win {predicted_2020_wins_for_team} games."


print(projected_wins_2020(chosen_team))

print("""
""")

def stat_improvement(chosen_team):
	matching_rows = nfl_df[(nfl_df["YEAR"] == 2019) & (nfl_df["Team"] == chosen_team)]

	DEF_TAKEAWAYS_GAME = matching_rows["DEF_TAKEAWAYS_GAME"]
	OFF_GIVEAWAYS_GAME = matching_rows["OFF_GIVEAWAYS_GAME"]
	DEF_SACKS_GAME = matching_rows["DEF_SACKS_GAME"]
	OPP_YARDS_RUSH = matching_rows["OPP_YARDS_RUSH"]
	OPP_YARDS_PASSATT = matching_rows["OPP_YARDS_PASSATT"]
	PASS_ATTEMPTS_GAME = matching_rows["PASS_ATTEMPTS_GAME"]
	RUSH_ATTEMPTS_GAME = matching_rows["RUSH_ATTEMPTS_GAME"]
	YARDS_PER_PASS_ATT = matching_rows["YARDS_PER_PASS_ATT"]

	chosen_stat = input("""What stat would you like to improve? Copy and paste one of the following:
	DEF_TAKEAWAYS_GAME
	OFF_GIVEAWAYS_GAME
	DEF_SACKS_GAME
	OPP_YARDS_RUSH
	OPP_YARDS_PASSATT
	PASS_ATTEMPTS_GAME
	RUSH_ATTEMPTS_GAME
	YARDS_PER_PASS_ATT
	
	Your choice: """)
 
	added_to_chosen_stat = float(input("""
	By how much would you like to improve that stat? Enter a numeric value only.
	Enter a negative value if you chose a stat where a lower number is better. 
	Your chosen value is: """))

	if chosen_stat == "DEF_TAKEAWAYS_GAME":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*(DEF_TAKEAWAYS_GAME + added_to_chosen_stat) + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	elif chosen_stat == "OFF_GIVEAWAYS_GAME":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*(OFF_GIVEAWAYS_GAME + added_to_chosen_stat) + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	elif chosen_stat == "DEF_SACKS_GAME":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*(DEF_SACKS_GAME + added_to_chosen_stat) + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	elif chosen_stat == "OPP_YARDS_RUSH":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*(OPP_YARDS_RUSH + added_to_chosen_stat) + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	elif chosen_stat == "OPP_YARDS_PASSATT":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*(OPP_YARDS_PASSATT + added_to_chosen_stat) + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	elif chosen_stat == "PASS_ATTEMPTS_GAME":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*(PASS_ATTEMPTS_GAME + added_to_chosen_stat) + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	elif chosen_stat == "RUSH_ATTEMPTS_GAME":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*(RUSH_ATTEMPTS_GAME + added_to_chosen_stat) + YARDS_PER_PASS_ATT_coefficient*YARDS_PER_PASS_ATT
	elif chosen_stat == "YARDS_PER_PASS_ATT":
		new_projected_wins_equation = intercept_coefficient + DEF_TAKEAWAYS_GAME_coefficient*DEF_TAKEAWAYS_GAME + OFF_GIVEAWAYS_GAME_coefficient*OFF_GIVEAWAYS_GAME + DEF_SACKS_GAME_coefficient*DEF_SACKS_GAME + OPP_YARDS_RUSH_coefficient*OPP_YARDS_RUSH + OPP_YARDS_PASSATT_coefficient*OPP_YARDS_PASSATT + PASS_ATTEMPTS_GAME_coefficient*PASS_ATTEMPTS_GAME + RUSH_ATTEMPTS_GAME_coefficient*RUSH_ATTEMPTS_GAME + YARDS_PER_PASS_ATT_coefficient*(YARDS_PER_PASS_ATT + added_to_chosen_stat)
	
	new_projected_wins_equation_values = new_projected_wins_equation.values

	return f"If you add {added_to_chosen_stat} to {chosen_stat}, your new projected win total for 2020 is {new_projected_wins_equation_values}."

print(stat_improvement(chosen_team))
```
